### Aufgabe 1

- Generate random stream with `random.randint(start,end)`
- interval [-10,10]

```python
from isda_streaming.data_stream import DataStream
import random

random.seed(84)
```

```python
# list
nums = [1,2,3,4,5]

# list to Datastream
ds = DataStream().from_collection(nums)

print(ds)
```

```python
ds = DataStream().from_collection(
    [random.randint(-10, 10) for _ in range (10)]
)

print(ds)
```

### Aufgabe 2

- Change 0 to -1 and 1

```python
def replace_zero(x):
    if x == 0:
        return [-1, 1]
    return [x]

replace_zero_ds = ds.flat_map(replace_zero)

print(replace_zero_ds)
```

### Aufgabe 3

- Split the stream into two streams: even and odd numbers

```python
def is_even(x):
    return x % 2 == 0

def is_odd(x):
    return x % 2 != 0

even_ds = replace_zero_ds.filter(is_even)
odd_ds = replace_zero_ds.filter(is_odd)

print(even_ds)
print(odd_ds)
```
