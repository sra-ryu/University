### Aufgabe 1.1

- make Verkauf table
- columns: id, produkt, region, preis
- rows: (1, 'Bier', 'Berlin', 3.00),
        (2, 'Limo', 'Berlin', 2.00),
        (3, 'Bier', 'München', 3.00),
        (4, 'Bier', 'Berlin', 3.00),
        (5, 'Limo', 'München', 2.00),
        (6, 'Wein', 'München', 5.00)

```python
import duckdb

con = duckdb.connect()

con.sql(
    """
    CREATE TABLE Verkauf (
        id UINTEGER PRIMARY KEY, 
        produkt TEXT, 
        region TEXT, 
        preis FLOAT
    )
    """
)

con.sql(
    """
    INSERT INTO Verkauf VALUES
        (1, 'Bier', 'Berlin', 3.00),
        (2, 'Limo', 'Berlin', 2.00),
        (3, 'Bier', 'München', 3.00),
        (4, 'Bier', 'Berlin', 3.00),
        (5, 'Limo', 'München', 2.00),
        (6, 'Wein', 'München', 5.00)
    """
)
```

### Aufgabe 1.2

- Calculate the total sales for Berlin

```python
con.sql(
    """
    SELECT SUM(preis) total_preis
    FROM Verkauf
    WHERE region = 'Berlin'
    """
)
```

### Aufgabe 1.3

- Calculate the total sales by regions

```python
con.sql(
    """
    SELECT region, SUM(preis) total_preis
    FROM Verkauf
    GROUP BY region
    """
)
```

### Aufgabe 1.4

- Count the number of sales for each product

```python
con.sql(
    """
    SELECT produkt, COUNT(*) count
    FROM Verkauf
    GROUP BY produkt
    """
)
```

### Aufgabe 1.5

- Calculate the total sales for Berlin

```python
con.sql(
    """
    SELECT produkt, SUM(preis)
    FROM Verkauf
    WHERE region = 'Berlin'
    GROUP BY produkt
    """
)
```

### Aufgabe 1.6

- Calculate the total sales by region and by product

```python
con.sql(
    """
    SELECT region, produkt, SUM(preis)
    FROM Verkauf
    GROUP BY produkt, region
    """
)
```

### Aufgabe 2.1

- make Verkauf and Kinde tables and save it as "sales.duckdb"
- columns of Verkauf: id, produkt, kunde_id, preis
- columns of Kunde: id, name, city
- rows of Verkauf: (1, 'Bier', 1, 3.00), (2, 'Limo', 2, 2.00), (3, 'Bier', 1, 3.00), (4, 'Wein', 3, 5.00)
- rows of Kunde: (1, 'Anna', 'Berlin'), (2, 'Bob', 'München'), (3, 'Clara', 'Berlin')

```python
con2 = duckdb.connect('sales.duckdb')

con2.sql(
    """
    CREATE TABLE Verkauf (
        id UINTEGER PRIMARY KEY, 
        produkt TEXT, 
        kunde_id UINTEGER,
        preis FLOAT
    )
    """
)

con2.sql(
    """
    CREATE TABLE Kunde (
        id UINTEGER PRIMARY KEY,
        name TEXT,
        city TEXT
    )
    """
)

con2.sql(
    """
    INSERT INTO Verkauf VALUES
        (1, 'Bier', 1, 3.00),
        (2, 'Limo', 2, 2.00),
        (3, 'Bier', 1, 3.00),
        (4, 'Wein', 3, 5.00)
    """
)

con2.sql(
    """
    INSERT INTO Kunde VALUES
        (1, 'Anna', 'Berlin'),
        (2, 'Bob', 'München'),
        (3, 'Clara', 'Berlin')
    """
)
```

```python
con2.sql("SELECT * FROM Verkauf")
```

```python
con2.sql("SELECT * FROM Kunde")
```

### Aufgabe 2.2

- Calculate the total sales by regions

```python
con2.sql(
    """
    SELECT city, SUM(preis)
    FROM Verkauf v JOIN Kunde k ON (v.kunde_id = k.id)
    GROUP BY city
    """
)
```

### Aufgabe 2.3

- Calculate the total sales by product for Berlin


```python
con2.sql(
    """
    SELECT produkt, SUM(preis)
    FROM Verkauf v JOIN Kunde k ON (v.kunde_id = k.id)
    WHERE city = 'Berlin'
    GROUP BY produkt
    """
)
```

### Aufgabe 2.4

- Find the product with total sales >= 5.00

```python
con2.sql(
    """
    SELECT v.produkt, SUM(preis)
    FROM Verkauf v JOIN Kunde k ON (v.kunde_id = k.id)
    GROUP BY produkt
    HAVING SUM(preis) >= 5.00
    """
)
```

### Aufgabe 2.5

- Calculate the total sales by city and product using ROLLUP

```python
con2.sql(
    """
    SELECT k.city, v.produkt, SUM(preis)
    FROM Verkauf v JOIN Kunde k ON (v.kunde_id = k.id)
    GROUP BY ROLLUP(city, produkt)
    """
)
```

### Aufgabe 2.6

- Calculate the total sales for all combinations of city and product

```python
con2.sql(
    """
    SELECT k.city, v.produkt, SUM(preis)
    FROM Verkauf v JOIN Kunde k ON (v.kunde_id = k.id)
    GROUP BY CUBE(city, produkt)
    """
)
```
