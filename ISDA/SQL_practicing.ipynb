### Aufgabe 1.1

- make Verkauf table
- columns: id, produkt, region, preis
- rows: (1, 'Bier', 'Berlin', 3.00),
        (2, 'Limo', 'Berlin', 2.00),
        (3, 'Bier', 'München', 3.00),
        (4, 'Bier', 'Berlin', 3.00),
        (5, 'Limo', 'München', 2.00),
        (6, 'Wein', 'München', 5.00)

```python
import duckdb

con = duckdb.connect()

con.sql(
    """
    CREATE TABLE Verkauf (
        id UINTEGER PRIMARY KEY, 
        produkt TEXT, 
        region TEXT, 
        preis FLOAT
    )
    """
)

con.sql(
    """
    INSERT INTO Verkauf VALUES
        (1, 'Bier', 'Berlin', 3.00),
        (2, 'Limo', 'Berlin', 2.00),
        (3, 'Bier', 'München', 3.00),
        (4, 'Bier', 'Berlin', 3.00),
        (5, 'Limo', 'München', 2.00),
        (6, 'Wein', 'München', 5.00)
    """
)
```

### Aufgabe 1.2

- Calculate the total sales for Berlin

```python
con.sql(
    """
    SELECT SUM(preis) total_preis
    FROM Verkauf
    WHERE region = 'Berlin'
    """
)
```

### Aufgabe 1.3

- Calculate the total sales by regions

```python
con.sql(
    """
    SELECT region, SUM(preis) total_preis
    FROM Verkauf
    GROUP BY region
    """
)
```

### Aufgabe 1.4

- Count the number of sales for each product

```python
con.sql(
    """
    SELECT produkt, COUNT(*) count
    FROM Verkauf
    GROUP BY produkt
    """
)
```

### Aufgabe 1.5

- Calculate the total sales for Berlin

```python
con.sql(
    """
    SELECT produkt, SUM(preis)
    FROM Verkauf
    WHERE region = 'Berlin'
    GROUP BY produkt
    """
)
```

### Aufgabe 1.6

- Calculate the total sales by region and by product

```python
con.sql(
    """
    SELECT region, produkt, SUM(preis)
    FROM Verkauf
    GROUP BY produkt, region
    """
)
```
