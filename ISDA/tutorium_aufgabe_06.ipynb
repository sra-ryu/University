## Aufgabe 3: Tabellen

- convert the ER-Diagramm into tables
- table:
    - Student(MatNr(PK), Name, Vorname, GebDatum, FakNr -> Fakultaet)
    - Pruefungsfach(PNr(PK), Fach, FakNr -> Fakultaet)
    - belegt(MatNr -> Student(PK), PNr -> Preufungsfach(PK), prueferin, Termin, Note)
    - Fakultaet(FakNr(PK), FakName, Dekan)

```python
con = duckdb.connect()
```

```python
# Erstellen die Tabelle Fakultaet
query = """
CREATE TABLE Fakultaet (
    FakNr BIGINT PRIMARY KEY,
    FakName VARCHAR(100),
    Dekan VARCHAR(50)
);
"""

con.sql(query)
```

```python
query = "SELECT * FROM Fakultaet"

con.sql(query)
```

```python
# Erstellen die Tabelle Pruefungsfach
query = """
CREATE TABLE Pruefungsfach(
    PNr BIGINT PRIMARY KEY,
    Fach VARCHAR(100),
    FakNr BIGINT NOT NULL REFERENCES Fakultaet(FakNr)
);
"""

con.sql(query)
```

```python
query = "SELECT * FROM Pruefungsfach"

con.sql(query)
```

```python
# Erstellen die Tabelle Student
query = """
CREATE TABLE Student(
    MatNr BIGINT PRIMARY KEY,
    Name VARCHAR(50),
    Vorname VARCHAR(50),
    GebDatum DATE,
    FakNr BIGINT NOT NULL REFERENCES Fakultaet(FakNr)
);
"""

con.sql(query)
```

```python
query = "SELECT * FROM Student"

con.sql(query)
```
