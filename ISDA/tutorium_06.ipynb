### 2. Öffnen einer Datenbankverbindung

- import duckdb and open the data bank.

```python
import duckdb

con = duckdb.connect()

# close duckdb
con.close()

con = duckdb.connect("intro.duckdb")
```

### 3. Erstellen einer Relation

- create a table
- table name: Person
- attributes: name(key), age

```python
con.sql("CREATE TABLE Person(Name TEXT PRIMARY KEY, Age UTINYINT);")
```

### 4. Löschen einer Relation

- delete the Person table
- create a Person table again, but the key is PersonId

```python
con.sql("DROP TABLE Person;")

con.sql("CREATE TABLE Person(PersonId UINTEGER PRIMARY KEY, Name TEXT, Age UTINYINT);")
```

### 5. Einfügen von Daten

- insert a new data
- PersonId = 1, Name = 'Peter Schmidt', Age = 23

```python
con.sql("INSERT INTO Person (PersonId, Name, Age) VALUES (1, 'Peter Schmidt', 23)")
```

- insert multiple data
- PersonId = 2, Name = 'Peter Schmidt', Age = 23
- PersonId = 3, Name = 'Sandra Müller', Age = 24
- PersonId = 4, Name = 'Petra Maier', Age = 14

```python
con.sql("INSERT INTO Person VALUES (2, 'Peter Schmidt', 23), (3, 'Sandra Müller', 24), (4, 'Petra Maier', 14)")
```

### 6. Ausgabe einer Relation

- querying and displaying data

```python
con.sql("SELECT * FROM Person")
```

### 7. Änderung (Updates) der Datenbank

- correct data
- change Age of Petra(14 -> 24)

```python
con.sql("UPDATE Person SET Age = Age + 10 WHERE PersonId = 4")
```

```python
# query all data from Person
con.sql("SELECT * FROM Person")
```
